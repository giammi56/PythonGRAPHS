```python
#original example from
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.griddata.html

import numpy as np
import itertools
from scipy.interpolate import griddata

x_rot=[]
y_rot=[]

def func(x, y):
    return (x-0.1)**2+(y+0.5)**2    

#create the MFPAD equivalent
#grid example
fixedth = np.around(np.array(list(itertools.product(np.linspace(-1,1,100).tolist(),np.linspace(-1,1,200).tolist()))),3)
xp = np.array([col[0] for col in fixedth]); #phi_full 20000
x = np.unique(xp)
yp = np.array([col[1] for col in fixedth]); #costheta_full 20000
y = np.unique(y)

points = np.array(list(zip(xp,yp)))

#line example
# xp = np.linspace(1,-1,1000)
# yp = np.linspace(1,-1,1000)
# points = np.array(list(zip(xp,yp)))

#random example
# points = np.random.rand(1000, 2)

#my original MFPAD
values = func(points[:,0], points[:,1])

#rotation of the full coordinates
for cth,ph in zip(yp.reshape(-1),xp.reshape(-1)):
    tempct=cth+0.3
    if tempct<-1.:
        y_rot.append(tempct+2.)
    elif tempct>1.:
        y_rot.append(tempct-2.)
    else:
        y_rot.append(tempct)

    tempp=ph+0.
    if tempp<-1.:
        x_rot.append(tempp+2.)
    elif tempp>1.:
        x_rot.append(tempp-2.)
    else:
        x_rot.append(tempp)

x_rot=np.array(x_rot)
y_rot=np.array(y_rot)

#unique coordiantes with the index i.e. not sorted
index_x = np.unique(x_rot, return_index=True)[1]
x_rot_uni = [x_rot[index] for index in sorted(index_x)]

index_y = np.unique(y_rot, return_index=True)[1]
y_rot_uni = [y_rot[index] for index in sorted(index_y)]

#rotated values: I can calculate them here, but I canÂ´t with my data! It is a check
values_rot = func(x_rot,y_rot)

#rotated meshgrid
grid_x, grid_y = np.meshgrid(x_rot_uni,y_rot_uni)

# grid_z0 = griddata(points, values, (grid_x, grid_y), method='nearest') #original with mgrid
grid_z0 = griddata(points, values, (grid_x.T, grid_y.T), method='nearest')

import matplotlib.pyplot as plt
plt.subplot(221)
# plt.imshow(func(grid_x, grid_y), extent=(0,1,0,1), origin='lower')
plt.imshow(values.reshape(100,200), extent=(-1,1,-1,1), origin='lower')
plt.title('Original')
plt.subplot(222)
plt.imshow(values_rot.reshape(100,200), extent=(-1,1,-1,1), origin='lower')
plt.title('rotated CALCULATED')
plt.subplot(223)
plt.tricontourf(grid_x.reshape(-1),grid_y.reshape(-1),grid_z0.reshape(-1), extent=(-1,1,-1,1), origin='lower')
plt.title('rotated countorf')
plt.subplot(224)
# plt.imshow(grid_z0.T, extent=(-180,180,-1,1), aspect = 180, origin='lower')
plt.imshow(grid_z0, extent=(-1,1,-1,1), origin='lower')
plt.title('rotated imgshow')
plt.gcf().set_size_inches(15, 15)
plt.show()
```

```python
from scipy.interpolate import griddata
import matplotlib.pyplot as plt
import itertools
import numpy as np

def func(x, y):
    return (x-0.1)**2+(y+0.5)**2
fixedth = np.around(np.array(list(itertools.product(np.linspace(-1,1,100).tolist(),np.linspace(-1,1,200).tolist()))),3)
xp = np.array([col[0] for col in fixedth]); #phi_full 20000
yp = np.array([col[1] for col in fixedth]); #costheta_full 20000
points = np.array(list(zip(xp,yp)))

values = func(points[:,0], points[:,1])

# grid_x, grid_y = np.mgrid[-1:1:100j, -1:1:200j]
grid_x, grid_y = np.mgrid[0:1:100j, 0:1:200j]
grid_z0 = griddata(points, values, (grid_x, grid_y), method='nearest')

plt.subplot(221)
# plt.imshow(func(grid_x, grid_y).T, extent=(0,1,0,1), origin='lower')
plt.imshow(values.reshape(100,200), extent=(-1,1,-1,1), origin='lower')
# plt.plot(points[:,0], points[:,1], 'k.', ms=1)
plt.title('Original')
plt.subplot(222)
plt.imshow(grid_z0.T, extent=(0,1,0,1), origin='lower')
plt.title('Nearest')

plt.gcf().set_size_inches(15, 15)

plt.show()
```
